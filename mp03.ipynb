```python
# Imports (all of them!)
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import os
import seaborn as sns
from sklearn.decomposition import TruncatedSVD
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
from glob import glob
from collections import defaultdict
from nltk import word_tokenize, sent_tokenize
import string
import pprint
import statistics as stat
from scipy import stats
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_validate
import warnings
warnings.filterwarnings('ignore')
```

```python
# Functions 

# Read lexicon, edited from ps_03
def read_lexicon(filepath):
    lexicon = defaultdict(dict)
    with open(filepath, 'r') as f:
    # lexicon file format is: word value
        for line in f:
            word, value = line.strip().split()
            # get rid of partofspeech
            word, part_of_speech = word.split("_",1)
            lexicon[word] = float(value)
    return lexicon

# Tokenizer from ps_03
def tokenize_text(input_text):
    return_sentences = []
    text_sentences = sent_tokenize(input_text)
    stopwords = []
    for punctuation in string.punctuation:
        stopwords.append(punctuation)
    for each_token in range(len(text_sentences)):
        text_sentences[each_token] = text_sentences[each_token].lower()
    for each_sentence in text_sentences:
        sentence_tokens = word_tokenize(each_sentence)
        return_sentences.append(sentence_tokens)
    return return_sentences

# Average inflammatory score for each novel
def inflammatory_avg(token_list, lexicon):
    check_list = []
    token_count = 0
    for novel_token_list in novel_tokens:
        for a_token in novel_token_list:
            token_count += 1
            key_check = a_token
            if key_check in lexicon.keys():
                check_list.append(lexicon[key_check])
        novel_sum = sum(check_list)
        novel_score = abs(novel_sum)/token_count
    return novel_score

# Compare scores function from mp_02
def compare_scores(scores_dict):
    '''
    Takes a dictionary of cross_validate scores.
    Returns a color-coded Pandas dataframe that summarizes those scores.
    '''
    import pandas as pd
    df = pd.DataFrame(scores_dict).T.applymap(np.mean).style.background_gradient(cmap='RdYlGn')
    return df

# Make labels function from mp_02
def make_labels(novel_tokens, keyword):
    labels = []
    is_it = 0
    for novel_token_list in novel_tokens:
        for a_token in novel_token_list:
            if a_token == keyword:
                is_it += 1
            else:
                is_it += 0
    if is_it > 0:
        labels.append(1)
    else:
        labels.append(0)
    return np.array(labels)
```

```python
# Read in novels and lexicon
metadata = pd.read_csv('class_corpus_metadata.csv')
relevant_novels = metadata
filename_list = os.listdir('class_corpus')
relevant_novels['file_here'] = relevant_novels['filename'].isin(filename_list)
relevant_novels = relevant_novels.loc[relevant_novels['file_here'] == True]

novels_path = glob(os.path.join('..', 'info3350', 'class_corpus', '*.txt'))
lex_path = os.path.join('..','info3350','data','lexicons','abuselex.txt')
abuselex = read_lexicon(lex_path)

# Inflammatory words count
inflammatory_dictionary = {}
for nov in novels_path:
    for text in relevant_novels['filename']:
        if '../info3350/class_corpus/'+text == nov:
            with open(nov, 'r', encoding = 'utf-8-sig') as f:
                novel = f.read()
                novel = novel.strip()
                novel_tokens = tokenize_text(novel)
                inflammatory_dictionary[text] = inflammatory_avg(novel_tokens, abuselex)

inflammatory_dataframe = pd.DataFrame([inflammatory_dictionary])
inflammatory_dataframe = inflammatory_dataframe.T
inflammatory_dataframe.index.name = 'filename'
inflammatory_dataframe.columns = ['inflammatory_scores']
relevant_novels = relevant_novels.merge(inflammatory_dataframe, how = 'inner', on = 'filename')
```
