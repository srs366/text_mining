First, in looking to see whether we are right about the number of female authors growing, from before the war to during/after it, I'd simply plot a stacked histogram, with bins spanning over every two years between 1789 and 1875, that maps how many books were written by male vs. female authors throughout these times. Starting with the tokenize_text function, I implemented this to remove extraneous characters from each novels' text, such as stopwords and punctuation, just because I thought it would make the analysis more revealing in differences between texts should we look at the tokens, specifically. 

```python
#import that which we'll need to tokenize
from nltk import word_tokenize, sent_tokenize
import string

#to tokenize the text
def tokenize_text(input_text):
    return_sentences = []
    text_sentences = sent_tokenize(input_text)
    stopwords = ['â€™','â€˜', '--', 'â€œ', 'â€']
    for punctuation in string.punctuation:
        stopwords.append(punctuation)
    for each_token in range(len(text_sentences)):
        text_sentences[each_token] = text_sentences[each_token].lower()
    for each_sentence in text_sentences:
        sentence_tokens = word_tokenize(each_sentence)
        filtered_sentence_tokens = [sentence_token for sentence_token in sentence_tokens if sentence_token not in stopwords]
        return_sentences.append(filtered_sentence_tokens)
    return return_sentences
```

```python
#gendered terms for our gender score function to consider in assigning gender to lists of tokens
male_words_revised = set([
    'guy','spokesman','chairman',"men's",'men','him',"he's",'his',
    'boy','boyfriend','boyfriends','boys','brother','brothers','dad',
    'dads','dude','father','fathers','fiance','gentleman','gentlemen',
    'god','grandfather','grandpa','grandson','groom','he','himself',
    'husband','husbands','king','male','man','mr','nephew','nephews',
    'priest','prince','son','sons','uncle','uncles','waiter','widower',
    'widowers', 'lad', 'lads', 'gay', 'masculine', 'fireman', 'policeman', 'milkman', 'duke'
])

female_words_revised = set([
    'heroine','spokeswoman','chairwoman',"women's",'actress','women',
    "she's",'her','aunt','aunts','bride','daughter','daughters','female',
    'fiancee','girl','girlfriend','girlfriends','girls','goddess',
    'granddaughter','grandma','grandmother','herself','ladies','lady',
    'lady','mom','moms','mother','mothers', 'mrs','ms','niece','nieces',
    'priestess','princess','queens','she','sister','sisters','waitress',
    'widow','widows','wife','wives','woman', 'lesbian', 'sapphic', 'feminine', 'duchess', 'madam', 'madame'
])

#our gender score function that returns a gendered score of a list of tokens, given the aforementioned gendered terms to consider
def gender_score(wl_token_list, male_wordlist, female_wordlist):
    binary_gender_score = 0
    male_binary_gender_score = 0
    female_binary_gender_score = 0
    for wl_token in wl_token_list:
        if wl_token in male_wordlist:
             male_binary_gender_score = male_binary_gender_score + 1
        elif wl_token in female_wordlist:
             female_binary_gender_score = female_binary_gender_score + 1
    if male_binary_gender_score > female_binary_gender_score:
        binary_gender_score = 0
    elif male_binary_gender_score < female_binary_gender_score:
        binary_gender_score = 1
    elif (male_binary_gender_score == female_binary_gender_score):
        binary_gender_score = None
    return binary_gender_score 
```

```python
# A function that will help read the inflammatory lexicon and parse it.
def read_lexicon(filepath):
    lexicon = defaultdict(dict)
    with open(filepath, 'r') as f:
    # lexicon file format is: word value
        for line in f:
            word, value = line.strip().split()
            # get rid of partofspeech
            word, part_of_speech = word.split("_",1)
            lexicon[word] = float(value)
    return lexicon
```

```python
# a function specific for reading in the emolex file, given that my other function isn't working on it.
def read_emolex(filepath=None):
    '''
    Takes a file path to the emolex lexicon file.
    Returns a dictionary of emolex sentiment values.
    '''
    if filepath==None: # Try to find the emolex file
        path = r'C:\Users\stefa_000\Desktop\INFO_3350\info3350-s22'
        filepath = os.path.join(path,'hw','emolex.txt')
        if os.path.isfile(filepath):
            pass
        elif os.path.isfile('emolex.txt'):
            filepath = 'emolex.txt'
        else:
            raise FileNotFoundError('No EmoLex file found')
    emolex = defaultdict(dict) # Like Counter(), defaultdict eases dictionary creation
    path = r'C:\Users\stefa_000\Desktop\INFO_3350\info3350-s22'
    filepath = os.path.join(path,'hw','emolex.txt')
    with open(filepath, 'r') as f:
    # emolex file format is: word emotion value
        for line in f:
            word, emotion, value = line.strip().split()
            emolex[word][emotion] = int(value)
    return emolex
```

```python
#captures sentiment score for a list of tokens utilizing lexicon to do so
def sentiment_score(sl_token_list, el_lexicon):
    sentiment_dictionary = defaultdict(int)
    for s_token in sl_token_list:
        if s_token in el_lexicon.keys():
            for specific_emotion in ['anger', 'anticipation', 'disgust', 'fear', 'joy', 'negative', 'positive', 'sadness', 'surprise', 'trust']:
                sentiment_dictionary[specific_emotion] = sentiment_dictionary[specific_emotion] + el_lexicon[s_token][specific_emotion]
    for specific_emotion in sentiment_dictionary:
        if len(sl_token_list) > 0:
            sentiment_dictionary[specific_emotion] = round(sentiment_dictionary[specific_emotion] / len(sl_token_list), 3)
    return sentiment_dictionary
```

Of course, we're also going to have to combine all of the following into one function so that we can properly gender a texts' tokens, attribute emotion to them, and then pair the gender and emotions to convey gendered emotions of its characters, which I've done below, once again borrowing from my code from PS03. 

```python
#function to calculate gendered sentiment
def gendered_sentiment(some_tokenized_text, male_words_input, female_words_input):
    male_sentiment_value = 0
    female_sentiment_value = 0
    gender_dict = dict({'male': 0, 'female': 0})
    final_dict = dict({'anger': gender_dict.copy(), 'anticipation': gender_dict.copy(), 'disgust': gender_dict.copy(), 'fear': gender_dict.copy(), 'joy': gender_dict.copy(), 'negative': gender_dict.copy(), 'positive': gender_dict.copy(), 'sadness': gender_dict.copy(), 'surprise': gender_dict.copy(), 'trust': gender_dict.copy()})
    
    for some_sentence in some_tokenized_text:
        some_gender_score = gender_score(some_sentence, male_words_input, female_words_input)
        some_sentiment_score = sentiment_score(some_sentence, emolex)
        if some_gender_score is not None:
            if some_gender_score == 0:
                male_sentiment_value = male_sentiment_value + 1
                for gs_key, gs_value in some_sentiment_score.items():
                    final_dict[gs_key]['male'] = final_dict[gs_key]['male'] + gs_value
            if some_gender_score == 1:
                female_sentiment_value = female_sentiment_value + 1
                for gs_key, gs_value in some_sentiment_score.items():
                    final_dict[gs_key]['female'] = final_dict[gs_key]['female'] + gs_value
                    
    for some_emotion, se_v in final_dict.items():
        if male_sentiment_value != 0:
            final_dict[some_emotion]['male']/= male_sentiment_value
        if female_sentiment_value != 0:
            final_dict[some_emotion]['female']/= female_sentiment_value 
    return final_dict
```
